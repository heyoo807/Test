```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
from collections import defaultdict

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id="60dda1b02dd84672a332036c4b81ac3f",
                                                           client_secret="98e1afbce10a43b4a8d01da78421289a"))

def find_song(name, year, artist, album):
    song_data = defaultdict()
    results = sp.search(q= 'album : {} track: {}  year: {}, artist : {} '.format(name, year, artist, album), limit=1)
    if results['tracks']['items'] == []:
        return None

    results = results['tracks']['items'][0]

    track_id =results['id']
    features = sp.audio_features(tracks=[track_id])[0]
    song_data['name'] = [name]
    song_data['album'] = results['album']['name']
    song_data['year'] = [year]
    song_data['artists'] = results['artists'][0]['name']
    song_data['release_date'] = int(results['album']['release_date'].split('-')[0])
    song_data['explicit'] = [int(results['explicit'])]
    song_data['duration_ms'] = [results['duration_ms']]
    song_data['popularity'] = [results['popularity']]
    song_data['genre'] = ' '
    if (song_data['release_date'] != year): # 가수명, 발매일 하나라도 다르면 저장 안함
      return
    if features is None: return
    # if (song_data['artists'] != artist): # 가수명, 발매일 하나라도 다르면 저장 안함
    #   return
    for key, value in features.items():
          if key in ['uri', 'track_href', 'analysis_url', 'type']: continue
          song_data[key] = value

    song_data = pd.DataFrame(song_data)
    return song_data[['valence', 'year', 'acousticness', 'artists', 'danceability', 'duration_ms', 'energy', 'explicit', 'id', 'instrumentalness', 'key', 'liveness', 'loudness', 'mode', 'name', 'album', 'genre', 'popularity', 'release_date', 'speechiness', 'tempo']]
```
